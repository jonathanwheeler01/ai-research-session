# Introduction to Generative AI
-------------------------------

5 minutes

**Questions**

- What exactly _is_ generative AI, and how is it different from other types of AI?
- What is a large language model (LLM)?
- How are LLMs different from search engines and databases?

**Objectives**

- Define generative AI as a research tool.
- Identify differences between generative AI tools and other research tools.


## Definitions

*Artificial intelligence* is a general term that includes many different technologies and applications. Our focus today is generative AI, which creates or generates content - chats, images, music, etc. - in response to a prompt. Generative AI is distinguished in this case from

- Machine learning: Leverages patterns in data to perform classifications and make predictions.
    - Supervised
        - Trained on data that have been manually pre-classified
    - Unsupervised
        - Data are classified algorithmically
- Neural networks: Layers of computing nodes that try to mimic the human brain
    - Deep learning: Machine learning with neural networks

All of these processes and applications play a role in the development of generative AI, and all are often included under the umbrella term **_AI_**. 

Large lanaguage models, or LLMs, are another important concept relative to ChatGPT and other language-based generative AI applications. Using the terms above in a somewhat non-exact way, LLMs can be considered as a type of unsupervised neural network that perform natural language processing on large text corpus (think big - like a snapshot of the entire publicly accessible web). Through statistical modeling of the corpus, LLMs predict the likelihood of one word following another word or phrase.

**Demonstration**

A good analogy is the word game some of us may have played, in which one person begins a story with a word, the next person adds a second word, the third person provides a third word, etc. For the most part, the farther along we get in the process the more limited the range of possible words becomes.

## Research with Generative AI

The structure of LLMs and the way in which the underlying data are processed makes them very different from the search engines and databases that we're used to using for research. To use a non-exact explanation:

- Databases index the features (metadata, full text) of articles, object, and other content. Search results are typically weighted or ranked based on similarities between indexed documents and a query.
- Generative AI generates a novel response to a prompt based on a model of the statistcal properties of large training corpora.

Keep in mind - generative AI tools are trained to respond to a prompt, and respond they will! A, AI response is not the same as a search result, and may not even be "information" as we tend to think of it. Whereas a database or search engine may return zero results to a query, ChatGPT and other generative AI tools are more likely to generate responses that are factually incorrect. 

**Demonstration**

1. Using two different web browsers, go to the [library website](https://library.unm.edu/) and [ChatGPT](https://chatgpt.com/) in both.
1. In the library search box in both browsers, search on *how to roast green chile*. Compare the results in the two browsers - they should be the same.
1. In the ChatGPT prompts in both browsers, ask *How do I roast green chile?* Compare the results - they are probably not the same!

What about a too narrow search?

1. From the library website, search for *green chile endorphin heat ratio*. No results - the search is either too narrow, or else no research has been done on this topic.
1. In ChatGPT, craft a prompt about endorphin/heat ratios in green chile. How would you characterize the response?

For a more cautionary example, consider citations as a type of structed text. Citations have standarized formats, which makes it makes it easy to come up with fake citations that look real. Below is ChatGPT's response to the prompt
